```python
import os

from google.colab import userdata

# Set your Groq API key
os.environ["GROQ_API_KEY"] = userdata.get('GROQ_API_KEY')
```

```python
from groq import Groq

client = Groq(
    api_key=os.environ.get("GROQ_API_KEY")
)
```

```python
def mentor_prompt(skill, level, user_input, history=""):
    return f"""
You are a Socratic mentor helping a student learn {skill} at {level} level.

Rules:
- Do NOT give direct answers immediately
- Ask guiding questions first
- Give hints step by step
- Encourage thinking
- Adjust difficulty based on response

Conversation so far:
{history}

User says:
{user_input}

Your response:
"""
```

```python
def mentor_chat(skill, level, user_input, history):
    prompt = mentor_prompt(skill, level, user_input, history)

    response = client.chat.completions.create(
        messages=[{"role": "user", "content": prompt}],
        model="llama3-70b-8192",  # fast + strong
        temperature=0.7
    )

    reply = response.choices[0].message.content
    return reply
```

```python
skill = input("Enter skill (e.g., Python, Math, Public Speaking): ")
level = input("Enter level (Beginner / Intermediate / Advanced): ")

history = ""

print("\nMentor: Let's start learning!\n")

while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit"]:
        print("Mentor: Keep practicing. You're improving.")
        break

    reply = mentor_chat(skill, level, user_input, history)

    print("\nMentor:", reply, "\n")

    # Update history
    history += f"\nUser: {user_input}\nMentor: {reply}\n"
```
